# **Latar Belakang**

Perkembangan dalam bidang teknologi informasi dan ilmu data membawa dampak signifikan, khususnya dalam hal adopsi metode komputasi untuk mendukung pengambilan keputusan berbasis data. Machine Learning berperan sebagai inti dari proses pengambilan keputusan yang berbasis data di berbagai sektor. Perkembangan mutakhir dalam teknologi informasi dan ilmu data telah memposisikan _Machine Learning_ (ML) sebagai disiplin ilmu esensial yang memfasilitasi pengambilan keputusan prediktif berbasis data historis (Santoso, 2021). ML memiliki kapabilitas untuk menyelesaikan dua tugas fundamental yaitu, klasifikasi untuk memprediksi label atau kategori diskret dan regresi untuk memprediksi nilai atau besaran kontinu (Han, Kamber, & Pei, 2012).

Pada pembelajaran ML, terdapat dua algoritma yang menjadi fokus utama dalam konteks pemodelan yaitu _decision tree_ (Pohon Keputusan) dan _K-Nearest Neighbors_ (KNN). _Decision Tree_ merupakan algoritma yang dikenal karena sifatnya yang transparan dan interpretabilitasnya yang superior, _Decision Tree_ memiliki kemampuan memvisualisasikan proses pengambilan keputusan. Secara fungsional, Decision Tree terbagi menjadi Classification Tree untuk prediksi kategori, dan Regression Tree yang dirancang secara spesifik untuk memprediksi nilai numerik kontinu (Ramadhani, 2020). Sementara itu, K-Nearest Neighbors (KNN) merupakan algoritma berbasis non-parametrik yang beroperasi berdasarkan prinsip kedekatan data dalam ruang fitur. Meskipun sering diaplikasikan untuk klasifikasi, prinsip ini secara efektif diekstensikan menjadi KNN Regresi (K-Nearest Neighbors Regression), di mana prediksi nilai kontinu bagi data baru dihitung berdasarkan rata-rata atau median nilai dari k tetangga terdekatnya (Nurhadi, 2019). Penguasaan varian regresi dari kedua algoritma ini (Regression Tree dan KNN Regresi) memberikan landasan metodologis yang kuat bagi mahasiswa untuk menangani spektrum luas permasalahan prediksi nilai di berbagai domain (Fadillah, 2021).

# _Feature Engineering_

## A - Missing Value

```python
# ============================================================
#  Cek missing values
# ============================================================
print(df.isnull().sum())
```

Setelah melewati proses sampling (n=1000), dataset penjualan BMW terbukti tidak mengandung *missing value*. Berdasarkan hasil pemeriksaan, semua variabel termasuk variabel target menunjukkan nilai nol (0) untuk data kosong, sehingga tidak diperlukan proses imputasi data lebih lanjut.

## B - Splitting Data

```python
# Tentukan target regresi
Y = df['Sales_Volume']

# Tentukan fitur (hapus Price_USD dari X)
X = df[['Year', 'Engine_Size_L', 'Mileage_KM', 'Price_USD']]
```

```python
# Split dataset (tanpa stratify karena regresi)
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.25, random_state=42
)

print("\nData Train:", X_train.shape)
print("Data Test:", X_test.shape)
```

Dataset dibagi menjadi dua bagian dengan proporsi 75% untuk data train dan 25% untuk data test, yang merupakan pembagian umum dalam pemodelan machine learning. Hasil pembagian menunjukkan bahwa data train berjumlah 750 baris dengan 4 fitur ((750, 4)), sedangkan data test berjumlah 250 baris dengan 4 fitur ((250, 4)). Data train digunakan untuk melatih model sehingga model dapat mempelajari pola dari dataset, sementara data test digunakan untuk mengevaluasi performa model terhadap data baru yang belum pernah dilihat sebelumnya. Pembagian ini dilakukan agar hasil evaluasi lebih objektif serta memastikan model memiliki kemampuan generalisasi yang baik dan tidak mengalami overfitting.

## C- Handling Outlier

```python
# ===========================
# D. HANDLING OUTLIER (IQR)
# ===========================

# Ambil hanya kolom numerik untuk perhitungan IQR
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns

# Hitung Q1, Q3, dan IQR untuk kolom numerik
Q1 = X_train[numeric_cols].quantile(0.25)
Q3 = X_train[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filtering: baris yang tidak outlier
filtered_index = ~(
    (X_train[numeric_cols] < (Q1 - 1.5 * IQR)) |
    (X_train[numeric_cols] > (Q3 + 1.5 * IQR))
).any(axis=1)

# Terapkan filter hanya pada X_train
X_train_filtered = X_train[filtered_index]

# Samakan y_train
Y_train_filtered = Y_train.loc[X_train_filtered.index]

# Reset index
X_train = X_train_filtered.reset_index(drop=True)
Y_train = Y_train_filtered.reset_index(drop=True)

print("\nJumlah data train setelah filtering outlier:", X_train.shape[0])
```

Pada tahap preprocessing, dilakukan penanganan outlier menggunakan metode *Interquartile Range* (IQR) untuk meningkatkan kualitas data sebelum proses pelatihan model. Berdasarkan hasil keluaran diperoleh jumlah data latih (*train* data) setelah proses *filtering outlier *adalah 750 observasi. Hal ini menunjukkan bahwa tidak ada *outlier* yang terdeteksi berdasarkan kriteria *Interquartile Range* (IQR) 1.5 * IQR. Sehingga, proses pemodelan dapat dilanjutkan dengan menggunakan seluruh 750 observasi, tanpa adanya penanganan outlier.

## D- Scaling Data

```python
# ===========================
# E. SCALING
# ===========================
# Identify numerical features directly from X_train
numerical_features = X_train.select_dtypes(include=np.number).columns.tolist()

# Initialize lists to store feature names based on skewness
nilai_skew = []
nilai_skew_normal = []

# Calculate skewness for each numerical feature in X_train
print("Calculating skewness for numerical features in X_train:")
for feature in numerical_features:
    skewness = X_train[feature].skew()
    print(f"  {feature}: {skewness:.4f}")
    if -0.5 <= skewness <= 0.5:
        nilai_skew_normal.append(feature)
    else:
        nilai_skew.append(feature)

# Display the categorized features
print('\nKolom yang mempunyai nilai skewness sedang (skewed):', nilai_skew)
print('Kolom yang mempunyai nilai skewness normal:', nilai_skew_normal)
```

Proses scaling dilakukan untuk memastikan bahwa seluruh variabel berada pada rentang skala yang sebanding. Berdasarkan *output*, hasil perhitungan skewness menunjukkan bahwa seluruh fitur numerik memiliki nilai dalam rentang -0.5 hingga 0.5. Nilai ini mengindikasikan bahwa distribusi data cenderung normal dan dikategorikan sebagai tidak skewed. Meskipun demikian, algoritma Machine Learning berbasis jarak seperti K-Nearest Neighbors (KNN) tetap sangat sensitif terhadap perbedaan skala antar fitur, sehingga proses standarisasi data tetap diperlukan.

```python
numerical_cols = ['Year', 'Engine_Size_L', 'Mileage_KM', 'Price_USD']
scaler = StandardScaler()

# Penerapan fit_transform pada X_train untuk kolom numerik
X_train_scaled = X_train.copy() # Create a copy to avoid modifying original X_train directly if not intended
X_train_scaled[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])

# Penerapan transform pada X_test untuk kolom numerik
X_test_scaled = X_test.copy() # Create a copy to avoid modifying original X_test directly if not intended
X_test_scaled[numerical_cols] = scaler.transform(X_test[numerical_cols])

print("Numerical features in X_train and X_test have been scaled using StandardScaler.")
print("X_train_scaled head:\n", X_train_scaled[numerical_cols].head())
print("X_test_scaled head:\n", X_test_scaled[numerical_cols].head())
```

# _Training Model_

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor()
model_knn.fit(X_train_scaled, Y_train)
```

```python
model_tree = DecisionTreeRegressor(random_state=42)
model_tree.fit(X_train_scaled, Y_train)
```

## _Hyperparameter Tunning_

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

# Buat model
model_knn = KNeighborsRegressor()
model_tree = DecisionTreeRegressor()

# Parameter KNN
param_knn = {"n_neighbors": range(3, 20)}

# Grid Search KNN
grid_knn = GridSearchCV(model_knn, param_knn, cv=5)
grid_knn.fit(X_train_scaled, Y_train)

# Parameter Decision Tree
param_tree = {
    "criterion": ["squared_error", "absolute_error", "friedman_mse"],
    "max_depth": range(2, 20)
}

# Grid Search Tree
grid_tree = GridSearchCV(model_tree, param_tree, cv=5)
grid_tree.fit(X_train_scaled, Y_train)

# Best model
best_knn = grid_knn.best_estimator_
best_tree = grid_tree.best_estimator_

print("Best KNN:", grid_knn.best_params_)
print("Best Decision Tree:", grid_tree.best_params_)

```

Hasil optimasi hyperparameter menggunakan Grid Search menunjukkan bahwa model K-Nearest Neighbors (KNN) mencapai kinerja terbaik dengan memilih 19 tetangga terdekat (*n_neighbors* = 19), yang mengindikasikan bahwa model cenderung menjadi lebih stabil dan robust terhadap noise dengan mengadopsi rata-rata dari cakupan data yang lebih luas, sehingga mengurangi varians. Sementara itu, model Regression Tree (Decision Tree Regressor) dioptimalkan dengan kriteria pemisahan '*squared_error*' dan kedalaman maksimum hanya 2 (*max_depth* = 2), yang menunjukkan bahwa struktur model yang sangat sederhana dan mudah diinterpretasikan sudah memadai untuk menangkap pola prediksi utama pada dataset penjualan BMW, sekaligus secara efektif mencegah *overfitting* yang sering terjadi pada pohon yang terlalu dalam.

# _Model Performance_

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

models = {
    "KNN": best_knn,
    "Decision Tree": best_tree
}

for name, model in models.items():
    print(f"\n===== {name} =====")
    y_pred = model.predict(X_test_scaled)

    mae = mean_absolute_error(Y_test, y_pred)
    mse = mean_squared_error(Y_test, y_pred)
    rmse = mean_squared_error(Y_test, y_pred) ** 0.5  # FIX
    r2 = r2_score(Y_test, y_pred)

    print("MAE :", mae)
    print("MSE :", mse)
    print("RMSE:", rmse)
    print("R² Score:", r2)
```

Berdasarkan hasil evaluasi model *performance* menggunakan metrik Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), dan $R^2$ Score, dapat disimpulkan bahwa Decision Tree Regressor menunjukkan kinerja yang sedikit lebih unggul dibandingkan dengan K-Nearest Neighbors (KNN) pada dataset penjualan BMW.

| Metrik                         | Decision Tree | K-Nearest Neighbors |
|--------------------------------|---------------|---------------------|
| Mean Absolute Error (MAE)      | 2617.46       | 2625.001             |
| Mean Squared Error (MSE)       | 8921657.02    | 9017787.04          |
| Root Mean Squared Error (RMSE) | 2986.91       | 3002.96             |
| R² Score                       | -0.0204       | -0.0314             |

Dalam konteks regresi (memprediksi volume penjualan), metrik MAE dan RMSE adalah yang paling penting karena mengukur kesalahan prediksi dalam unit penjualan. Decision Tree menghasilkan nilai MAE dan RMSE yang lebih rendah, menunjukkan bahwa kesalahan rata-rata prediksinya lebih kecil. Meskipun R² Score keduanya negatif (mengindikasikan bahwa kedua model gagal menangkap pola signifikan), Decision Tree berada lebih dekat ke nol (lebih baik daripada KNN).
Dengan demikian, Decision Tree Regressor direkomendasikan sebagai model yang lebih optimal di antara dua algoritma yang dibandingkan untuk memprediksi volume penjualan BMW, karena secara konsisten menghasilkan error prediksi yang lebih rendah

# Kesimpulan

Berdasarkan perbandingan metrik evaluasi, algoritma Decision Tree Regressor menunjukkan kinerja yang lebih unggul karena secara konsisten menghasilkan kesalahan prediksi yang lebih rendah (MAE 2617.46 dan RMSE 2986.91) dibandingkan dengan K-Nearest Neighbors (KNN). Meskipun nilai $R^2$ Score keduanya negatif—mengindikasikan bahwa model gagal menangkap variabilitas data secara memuaskan—Decision Tree tetap lebih mendekati nol (lebih baik) daripada KNN.
